# Example Visualization using OpenStreetMap export

```python
# Libraries
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(22, 12))

# Shapefiles
#   Exported from: http://www.openstreetmap.org/export#map=16/51.3416/5.2498
#   Converted using: https://geoconverter.hsr.ch
#   And unzipped
location_shapefile = 'borders/NLD_adm_shp/NLD_adm1'
location_shapefile_OSM_lines = 'openstreetmap/map.osm/lines'
location_shapefile_OSM_multilinestrings = 'openstreetmap/map.osm/multilinestrings'
location_shapefile_OSM_multipolygons = 'openstreetmap/map.osm/multipolygons'
location_shapefile_OSM_points = 'openstreetmap/map.osm/points'

# Libraries for mapping
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

map = Basemap(llcrnrlat=51.33,urcrnrlat=51.35,llcrnrlon=5.24,urcrnrlon=5.27, projection='tmerc', lat_0=52, lon_0=5, 
              resolution = 'h', area_thresh = 0.0001)

map.drawmapboundary(fill_color='aqua')
map.fillcontinents(color='#ddaa66',lake_color='aqua')
map.drawcoastlines()
map.drawcountries()

map.readshapefile(location_shapefile_OSM_lines, 'lines')
map.readshapefile(location_shapefile_OSM_multipolygons, 'multipolygons')
map.readshapefile(location_shapefile_OSM_points, 'points')

plt.show()
```
